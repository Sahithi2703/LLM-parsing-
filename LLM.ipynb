```python
from transformers import pipeline
import re
import json

model_name = "gpt2"
generator = pipeline('text-generation', model=model_name)
prompt = "Alice ordered 3 pizzas from Domino's on 15th August 2024."
output = generator(prompt, max_length=100, num_return_sequences=1)
print("Generated Output:")
print(output[0]['generated_text'])
parsed_output = output[0]['generated_text'].split('. ')

def extract_info(text):
    """Extracts name, pizza count, company, and date from the given text."""
    name_pattern = r"(\w+) ordered"
    pizza_count_pattern = r"(\d+) pizzas"
    company_pattern = r"from ([\w']+)"
    date_pattern = r"on (\d{1,2}[a-z]{2} \w+ \d{4})"

    name_match = re.search(name_pattern, text)
    pizza_count_match = re.search(pizza_count_pattern, text)
    company_match = re.search(company_pattern, text)
    date_match = re.search(date_pattern, text)

    date_formatted = date_match.group(1) if date_match else None

    extracted_data = {
        "name": name_match.group(1) if name_match else None,
        "pizza_count": int(pizza_count_match.group(1)) if pizza_count_match else None,
        "company": company_match.group(1) if company_match else None,
        "date": date_formatted
    }

    return extracted_data

parsed_info = extract_info(output[0]['generated_text'])
print("\nParsed Output:")
parsed_output= json.dumps(parsed_info, indent=4)
print(parsed_output) # Print the JSON directly for better readability
```
